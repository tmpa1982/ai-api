```python
from langchain_community.document_loaders import PyPDFLoader, WebBaseLoader
import bs4
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_openai import OpenAIEmbeddings

website_path = "https://www.networxrecruitment.com/Jobs/Advert/3993581?js=414&LinkType=1&key=2dqrwZQX2lwyteZYYcf3vub5WNjmfRjtZk3VspqAYjk%3D"

bs4_strainer = bs4.SoupStrainer(class_=("post-title", "post-header", "post-content"))
loader = WebBaseLoader(
    web_paths=(website_path,),
    # bs_kwargs={"parse_only": bs4_strainer},
)
website_docs = loader.load()
print(website_docs[0].page_content[:500])
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,  # chunk size (characters)
    chunk_overlap=200,  # chunk overlap (characters)
    add_start_index=True,  # track index in original document
)
website_all_splits = text_splitter.split_documents(website_docs)
```

```python
for content in website_all_splits:
    print(content.page_content)
```
